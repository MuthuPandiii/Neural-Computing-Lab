```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
X = np.arange(0,5.05,0.1)
mfx = fuzz.trapmf(X,[2,2.5,3,4.5])
```

```python
defuzz_centroid = fuzz.defuzz(X,mfx,'centroid')
defuzz_bisector = fuzz.defuzz(X,mfx,'bisector')
defuzz_mom = fuzz.defuzz(X,mfx,'mom')
defuzz_som = fuzz.defuzz(X,mfx,'som')
defuzz_lom = fuzz.defuzz(X,mfx,'lom')
```

```python
x_vals = [defuzz_bisector,defuzz_centroid,defuzz_mom,defuzz_som,defuzz_lom]
labels = ['Centroid','bisector','mean of maximum','min of maximum','max of maximun']
```

```python
y_max = [fuzz.interp_membership(X,mfx,i) for i in x_vals]
```

```python
colors = ['r','g','b','c','m']
```

```python
plt.figure(figsize=(14,7))

plt.plot(X,mfx,'k')
for x,y,color,label in zip(x_vals,y_max,colors,labels):
    plt.vlines(x,0,y,color,label=label)
plt.legend()
plt.show()
```
