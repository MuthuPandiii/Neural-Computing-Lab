```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class RBF:
    def __init__(self,n_input,n_hidden,n_output,sigma,random_state):
        self.n_input = n_input
        self.n_hidden = n_hidden
        self.n_output = n_output
        self.gamma = 1 / (2*sigma**2)
        self.random_state = random_state
        self.W = np.random.random((n_output,n_hidden))
    
    def rbf_kernel(self,x,c):
        return np.exp(-self.gamma * np.linalg.norm(x-c,axis=-1)**2)
    
    def activation(self,X):
        return self.rbf_kernel(self.hidden[:,np.newaxis,:],X[np.newaxis,:,:])
    
    def train(self,X,Y):
        if self.random_state != None:
            np.random.seed(self.random_state)
        rnd_idx = np.random.choice(X.shape[0],self.n_hidden)
        self.hidden = X[rnd_idx]
        
        self.W = Y.T @ np.linalg.pinv(self.activation(X))
        
    def predict(self,X):
        return (self.W @ self.activation(X)).T

X = np.linspace(-1,1,100).reshape(-1,1)
Y = np.sin(3 * (X**3) + 0.5)
Y+= np.random.normal(0,0.1,Y.shape)

rbf = RBF(1,10,1,1,4)

rbf.train(X,Y)
y_pred = rbf.predict(X)

plt.figure(figsize=(14,7))
plt.plot(X,Y,'k',label='Original Signal')
plt.plot(X,y_pred,'r',lw = 2,label='RBF Signal')
plt.title('RBF Regression')
plt.legend()
plt.grid()
plt.show()
```
